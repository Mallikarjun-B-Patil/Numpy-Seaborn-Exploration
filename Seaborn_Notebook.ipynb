# **DATA VISUALIZATION-SEABORN**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dist Plots

```python
# Histogram Plot

sns.histplot([0, 1, 2, 3, 4, 5])

plt.show()
```

```python
# Distribution Plot

sns.distplot([0, 1, 2, 3, 4, 5])

plt.show()
```

```python
# Distribution Plot without Histogram

sns.distplot([0, 1, 2, 3, 4, 5], hist=False)

plt.show()
```

# Normal Distribution

```python
# Random Normal array of Size

arr = np.random.normal(size=(2, 3))
arr
```

```python
# Random Normal array of Mean, SD, Size

arr = np.random.normal(loc=1, scale=2, size=100)
arr
```

```python
# Plot of 1000 values

sns.distplot(np.random.normal(size=1000))

plt.show()
```

```python
# Without histogram

arr = np.random.normal(size=1000)
sns.distplot(arr, hist=False)

plt.show()
```

# Binomial Distribution

```python
# Binomial Distribution of Rolling Dice

arr = np.random.binomial(n= 6, p= 0.5, size= 100)
arr
```

```python
# Binomial Distribution (n, p, size)

arr = np.random.binomial(n= 10, p= 0.5, size= 100)
sns.distplot(arr)

plt.show()
```

```python
# Combination of Normal and Binomial Distribution

arr_1 = np.random.normal(loc= 50, scale= 5, size= 100)
arr_2 = np.random.binomial(n= 100, p= 0.5, size= 100)

sns.distplot(arr_1, hist= False, label= "Normal")
sns.distplot(arr_2, hist= False, label= "Binomial")

plt.legend()
plt.show()
```

# Poisson Distribution

```python
arr = np.random.poisson(lam=2, size=100)
arr
```

```python
arr = np.random.poisson(lam=2, size=100)
sns.distplot(arr)

plt.plot()
```

```python
# Difference between Normal, Binomial and Poisson Distribution
# n*p = lam (Approx)
# n*p = loc (Approx)

sns.distplot(np.random.normal(loc=10, scale=3, size=1000), hist=False, label='Normal')
sns.distplot(np.random.binomial(n=1000, p=0.01, size=1000), hist=False, label='Binomial')
sns.distplot(np.random.poisson(lam=10, size=1000), hist=False, label='Poisson')

plt.legend()
plt.show()
```

# Uniform Distribution

```python
# Uniform array (0,1) - Has equal number of chances of occuring

arr = np.random.uniform(size=100)
arr
```

```python
# Uniform Distribution Plot

arr = np.random.uniform(size=1000)
sns.distplot(arr)

plt.plot()
```

# Logistic Distribution

```python
# Normal Distribution

arr = np.random.normal(loc=1, scale=2, size=10)
arr
```

```python
# Logistic Distribution

arr = np.random.logistic(loc=1, scale=2, size=10)
arr
```

```python
# Logistic Distribution Plot

arr = np.random.logistic(size=1000)
sns.distplot(arr)

plt.plot()
```

```python
# Difference between Normal and Logistic Distribution

sns.distplot(np.random.normal(size=1000), hist=False, label='Normal')
sns.distplot(np.random.logistic(size=1000), hist=False, label='Logistic')

plt.legend()
plt.show()
```
